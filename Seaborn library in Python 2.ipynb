```python
# Bar plot in seaborn library
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df1=sns.load_dataset('penguins')
```

```python
df1
```

```python
df1['island']
```

```python
sns.barplot(x=df1.island , y= df1.bill_length_mm)
plt.show()
```

```python
sns.barplot(x='island',y='bill_length_mm',data=df1)
```

```python
df1
```

```python
sns.barplot(x='island',y='bill_length_mm',data=df1,hue='sex')
plt.show()
```

```python
# Order parameter
```

```python
order1= ['Biscoe','Dream','Torgersen']
sns.barplot(x='island', y='bill_length_mm',data=df1,hue='sex',order=order1)
plt.show()
```

```python
# hue_order
```

```python
order1= ['Biscoe','Dream','Torgersen']
sns.barplot(x='island', y='bill_length_mm' , data=df1, hue='sex', order=order1, hue_order=['Female','Male'])
plt.show()
```

```python
order1=['Biscoe','Dream','Torgersen']
sns.barplot(x='island', y='bill_length_mm', data=df1, order=order1, color='green')
plt.show()
```

```python
order1=['Biscoe', 'Dream', 'Torgersen']
sns.barplot(x='island', y='bill_length_mm', data=df1, order=order1,hue='sex', palette='BuGn')
plt.show()
```

```python
order1=['Biscoe','Dream','Torgersen']
sns.barplot(x='island',y='bill_length_mm', data=df1, order=order1, hue='sex', palette='CMRmap')
plt.show()
```

```python
order1=['Biscoe','Dream','Torgersen']
sns.barplot(x='island',y='bill_length_mm',data=df1,order=order1,hue='sex',palette='Set1')
plt.show()
```

```python
#saturation -- value should be between 0 to 1
```

```python
order=['Biscoe','Dream','Torgersen']
sns.barplot(x='island',y='bill_length_mm',data=df1,hue='sex', order=order1,palette='Set1',saturation=1)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
sns.displot(df1['flipper_length_mm'])
plt.show()
```

```python
#bins
```

```python
sns.displot(df1['flipper_length_mm'],bins=[170,180,190,200,210,220,230,240])
plt.show()
```

```python
# change color
```

```python
sns.displot(df1['flipper_length_mm'], bins=[170,180,190,200,210,220,230,240], color='red')
plt.show()
```

```python
# kde
```

```python
sns.displot(df1['flipper_length_mm'],bins=[170,180,190,200,210,220,230,240],kde=True,color='g')
plt.show()
```
